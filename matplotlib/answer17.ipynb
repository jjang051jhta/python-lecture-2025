### 1) 승선 항구(`Embarked`)별 **전체 승객 수**를 계산하여 세로 막대그래프로 시각화하시오.

조건

- x축: Embarked (C, Q, S)
- y축: 승객 수
- 결측치 제거
- 막대 위에 값 표시

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"   # 윈도우 기본 한글 폰트
plt.rcParams["axes.unicode_minus"] = False  
titanic = pd.read_csv("./titanic.csv")
#print(titanic)
embarked_count = titanic["Embarked"].dropna().value_counts()
print(embarked_count.index)
print(embarked_count.values)
plt.bar(embarked_count.index,embarked_count.values, 
        color=["#ebb400","#00ade2","#02f00e"], 
        edgecolor="#000000")
plt.ylim(0,embarked_count.max()*1.15)
for i,value in enumerate(embarked_count.values):
  plt.text(i,value+5, value,ha="center", va="bottom")
plt.title("항구별 승선인원")
plt.xlabel("항구이름")
plt.ylabel("인원수")
```

### 2)연령(`Age`)을 기준으로 **연령대별 평균 생존 비율**을 계산하여 선형 그래프로 시각화하시오.

조건

- 연령 구간 예: 0–9, 10–19, 20–29 …
- 결측치 제거
- marker 표시

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"   # 윈도우 기본 한글 폰트
plt.rcParams["axes.unicode_minus"] = False  
titanic = pd.read_csv("./titanic.csv")
#print(titanic)
#print(titanic["Age"].isna().sum()) #NaN 데이터가 들어가 있는 갯수
df = titanic.dropna(subset=["Age"]).copy()  #Age에 NaN이 있으면 제거  다른 컬럼에 NaN이 있어도 제거하지 마라
#print(df["Age"].isna().sum())
df["AgeGroup"] = pd.cut(df["Age"], bins=range(0,81,10))
age_survive = df.groupby("AgeGroup")["Survived"].mean()
plt.plot(age_survive.index.astype(str),age_survive.values, marker="o")

#print(df)


```

### 3) 생존 여부(`Survived`)에 따른

**평균 요금(Fare)** 을 비교하여 막대그래프로 시각화하시오.

조건

- x축: Survived (0, 1)
- y축: 평균 Fare
- 막대 위에 평균 요금 표시

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"   # 윈도우 기본 한글 폰트
plt.rcParams["axes.unicode_minus"] = False  
titanic = pd.read_csv("./titanic.csv")

fare_mean =  titanic.groupby("Survived")["Fare"].mean()
#print(fare_mean)
#print(fare_mean.index)
plt.bar(fare_mean.index.astype(str), fare_mean.values)
for idx, value  in enumerate(fare_mean.values):
  plt.text(idx,value+fare_mean.max()*0.05, round(value,1), ha="center", va="bottom")
plt.ylim(0,fare_mean.max()+20)
plt.xlabel("생존자/사망자")
plt.ylabel("요금")
plt.title("생존 여부별 평균 요금")
```

### **4) 선형 그래프**

승선 항구(`Embarked`)별 **평균 요금(Fare)** 을 계산하여 선형 그래프로 시각화하시오.

조건

- y축: 평균 Fare

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"   # 윈도우 기본 한글 폰트
plt.rcParams["axes.unicode_minus"] = False  
titanic = pd.read_csv("./titanic.csv")
print(titanic["Embarked"].isna().sum())  #결측치 확인
print(titanic["Embarked"].isna().any())  #결측치 확인
fare_embarked = titanic.dropna(subset=["Embarked"]).groupby("Embarked")["Fare"].mean()
plt.plot(fare_embarked.index, fare_embarked.values, marker="o")
plt.xlabel("승선지")
plt.ylabel("승선항구별 평균요금")
plt.grid(True)
```

### **5)막대그래프**

연령(`Age`)을 기준으로 **연령대별 전체 승객 수**를 계산하여 막대그래프로 시각화하시오.

조건

- 연령 구간 직접 생성
- 결측치 제거
- 막대 색상 자유

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"   # 윈도우 기본 한글 폰트
plt.rcParams["axes.unicode_minus"] = False  
titanic = pd.read_csv("./titanic.csv")
print(titanic["Age"].isna().sum())
df = titanic.dropna(subset=["Age"])
df["AgeGroup"] = pd.cut(df["Age"], bins = range(0,80,10))
#print(df["AgeGroup"].value_counts().sort_index())
# (0, 10]  0 < 값 <= 10  10세 이하 
age_count = df["AgeGroup"].value_counts().sort_index()
#print(age_count.index)
labels = [ f"{int(interval.left)}~{int(interval.right-1)}" 
            for interval in age_count.index
         ]
#plt.bar(age_count.index.astype(str), age_count.values)
plt.bar(labels, age_count.values)
plt.xlabel("연령별")
plt.ylabel("연령별 승객수")
#plt.xticks(rotation=45)
```

### **6)선형 그래프**

요금(`Fare`) 구간별 **생존 비율 변화**를 선형 그래프로 시각화하시오.

조건

- Fare 구간 예: 0–50, 50–100, 100–200 …
- y축: 생존 비율
- marker 표시

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"   # 윈도우 기본 한글 폰트
plt.rcParams["axes.unicode_minus"] = False  
titanic = pd.read_csv("./titanic.csv")

titanic["FareGroup"]= pd.cut(titanic["Fare"], bins=[0,50,100,200,300,600])
fare_survived= titanic.groupby("FareGroup")["Survived"].mean()
#print(fare_survived.index)
labels= [f"{int(interval.left)}~{int(interval.right)}" for interval in fare_survived.index]
plt.plot(labels,fare_survived.values,marker="o" )
#plt.plot(fare_survived.index.astype(str),fare_survived.values,marker="o" )
plt.xlabel("승선 요금 구간")
plt.ylabel("생존율")
plt.title("요금 구간별 생존 비율")
```

### **7)막대그래프**

이름(`Name`)에서 호칭(Title: Mr, Mrs, Miss 등)을 추출하여 **호칭별 승객 수**를 막대그래프로 시각화하시오.

조건

- Title 파생 컬럼 생성
- 상위 4개 호칭만 사용
- 막대 위에 값 표시

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"   # 윈도우 기본 한글 폰트
plt.rcParams["axes.unicode_minus"] = False  
titanic = pd.read_csv("./titanic.csv")

titanic["Title"] = titanic["Name"].str.extract(" ([A-Za-z]+)\.") # 공백다음에 영어가 나오고 마지막에 .이 나오는 글자 추출
#print(titanic["Title"])
title_count = titanic["Title"].value_counts().head(4)
plt.bar(title_count.index, title_count.values)
for idx, value in enumerate(title_count.values):
  plt.text(idx,value+4, value, ha="center")

```
