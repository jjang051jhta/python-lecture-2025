### 1) 생존 여부 비율 파이차트

`Survived` 컬럼을 이용해 사망자(0)와 생존자(1)의 비율을 파이차트로 시각화하시오.

조건

- 라벨: Dead, Survived
- 퍼센트 표시
- 시작 각도는 90도

````python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic = pd.read_csv("./titanic.csv")

#print(titanic["Survived"].value_counts())
survived_count = titanic["Survived"].value_counts().sort_index()
plt.pie(
  survived_count,
  labels=["사망자","생존자"],
  autopct="%0.1f%%",
  startangle=90,
  shadow=True,
  explode=(0,0.06),
  # wedgeprops={
  #   "edgecolor":"#ffffff",
  #   "linewidth":2
  # }
)
plt.title("생존비율 표시")

````

### 2) 성별 생존 비율 파이차트

`Sex`와 `Survived` 컬럼을 이용하여 남성과 여성 각각의 생존 비율을 파이차트로 시각화하시오.

조건

- 남성과 여성을 각각 하나의 파이차트로 표현
- 퍼센트 표시

````python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic = pd.read_csv("./titanic.csv")

sex_list=["male","female"]

#print(titanic["Survived"].value_counts())
for idx, sex in enumerate(sex_list,start=1):
  sub =  titanic[titanic["Sex"]== sex]
  survived_count = sub["Survived"].value_counts().sort_index()
  plt.subplot(1,2,idx)  # 1행 2열
  plt.pie(
    survived_count,
    labels=["사망자","생존자"],
    autopct="%0.1f%%",
    startangle=90,
    shadow=True,
    explode=(0,0.06),
    
  )
  plt.tight_layout()
  plt.title(f"{sex}생존비율 표시")

````

### 3) 객실 등급별(Pclass) 생존자 비율 파이차트

`Pclass` 컬럼을 기준으로 각 객실 등급별 생존자 비율을 파이차트로 나타내시오.

조건

- 1등석, 2등석, 3등석 구분
- 생존자만을 기준으로 비율 계산

````python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic = pd.read_csv("./titanic.csv")

survivied =  titanic[titanic["Survived"]==1]
pclass_ratio =  survivied["Pclass"].value_counts().sort_index()
#print(pclass_survivied)
print(pclass_ratio)

plt.pie(
    pclass_ratio,
    labels=["1st","2nd","3rd"],
    autopct="%0.1f%%",
    startangle=90,
    shadow=True,
    explode=(0.06,0,0),
  )

#plt.tight_layout()
plt.title(f"class별 생존자 비율 표시")

````

### 4) 나이(Age) 히스토그램

`Age` 컬럼을 이용해 승객 나이 분포를 히스토그램으로 시각화하시오.

조건

- 결측치는 제거
- bin 개수는 20
- x축: Age, y축: Count

````python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic = pd.read_csv("./titanic.csv")

age = titanic["Age"].dropna()
print(age)
age_clip = age.clip(upper=80)  # 80세 이상을 80세에 몰아서 표시
plt.hist(age_clip,bins=10, 
         range=(0,80),  # 80세 미만  80세 이상데이터는 표시되지 않는다.
         rwidth=0.9)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("나이분포 히스토그램")
````

### 7) 나이(Age) × 요금(Fare) 산점도

`Age`와 `Fare`를 이용해 두 변수의 관계를 산점도로 나타내시오.

조건

- 색상은 `Survived` 기준
- 투명도(alpha) 적용

---

````python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic = pd.read_csv("./titanic.csv")
cmap = ListedColormap(["#bb00ab","#3be607"])

temp = titanic[["Age","Fare","Survived"]].dropna()
plt.scatter(temp["Age"], temp["Fare"], c=temp["Survived"], cmap=cmap, alpha=0.5)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.colorbar(ticks=[0,1],label="Survived (0 = Dead , 1 = Survived)")
plt.title("Age x Fare")
````

### 8) 나이(Age) × 가족 수(FamilySize) 산점도

다음 공식을 이용해 `FamilySize` 컬럼을 생성하시오.

```
FamilySize = SibSp + Parch +1

```

이후

Age와 FamilySize의 관계를 산점도로 시각화하시오.

조건

- 생존 여부에 따라 색상 구분

````python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic = pd.read_csv("./titanic.csv")
titanic02 = titanic.copy()
titanic02["FamilySize"] = titanic02["SibSp"]+titanic02["Parch"] + 1 # (본인 포함하기 위해 +1 을 해야함) 

temp = titanic02[["Age","FamilySize","Survived"]].dropna()
cmap = ListedColormap(["#bb00ab","#3be607"])

plt.scatter(temp["Age"], temp["FamilySize"], c=temp["Survived"], cmap=cmap, alpha=0.5)
plt.xlabel("Age")
plt.ylabel("FamilySize")
plt.colorbar(ticks=[0,1],label="Survived (0 = Dead , 1 = Survived)")
plt.title("Age x FamilySize (color = Survived)")
````

### 9) 요금(Fare) 기준 누적 평균 생존율 산점도

`Fare` 기준으로 데이터를 정렬한 뒤 누적 평균 생존율을 계산하고 산점도로 표현하시오.

조건

- x축: 승객 순서(Fare 기준)
- y축: 누적 평균 생존율

````python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic = pd.read_csv("./titanic.csv")

temp = titanic[["Fare","Survived"]].dropna().sort_values("Fare").reset_index(drop=True)
temp["cum_survived_rate"]= temp["Survived"].expanding().mean()
# Fare,Survived 둘중 하나라도 결측이 있으면 삭제 Fare를 기준으로 올림차순 기존 인덱스를 지우고 index를 새로 설정
print(temp)
plt.scatter(temp.index, temp["cum_survived_rate"], alpha=0.5, s=20, edgecolor="#000000", linewidths=0.2)
plt.xlabel("승객")
plt.ylabel("누적 평균 생존자 비율")
plt.title("요금 절렬후 누적평균 생존율")
````
