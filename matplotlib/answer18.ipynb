### 1) 생존 여부 비율 파이차트

`Survived` 컬럼을 이용해 사망자(0)와 생존자(1)의 비율을 파이차트로 시각화하시오.

조건

- 라벨: Dead, Survived
- 퍼센트 표시
- 시작 각도는 90도

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic = pd.read_csv("./titanic.csv")

#print(titanic["Survived"].value_counts())
survived_count = titanic["Survived"].value_counts().sort_index()
plt.pie(
  survived_count,
  labels=["사망자","생존자"],
  autopct="%0.1f%%",
  startangle=90,
  shadow=True,
  explode=(0,0.06),
  # wedgeprops={
  #   "edgecolor":"#ffffff",
  #   "linewidth":2
  # }
)
plt.title("생존비율 표시")

```

### 2) 성별 생존 비율 파이차트

`Sex`와 `Survived` 컬럼을 이용하여 남성과 여성 각각의 생존 비율을 파이차트로 시각화하시오.

조건

- 남성과 여성을 각각 하나의 파이차트로 표현
- 퍼센트 표시

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic = pd.read_csv("./titanic.csv")

sex_list=["male","female"]

#print(titanic["Survived"].value_counts())
for idx, sex in enumerate(sex_list,start=1):
  sub =  titanic[titanic["Sex"]== sex]
  survived_count = sub["Survived"].value_counts().sort_index()
  plt.subplot(1,2,idx)  # 1행 2열
  plt.pie(
    survived_count,
    labels=["사망자","생존자"],
    autopct="%0.1f%%",
    startangle=90,
    shadow=True,
    explode=(0,0.06),
    
  )
  plt.tight_layout()
  plt.title(f"{sex}생존비율 표시")

```

### 3) 객실 등급별(Pclass) 생존자 비율 파이차트

`Pclass` 컬럼을 기준으로 각 객실 등급별 생존자 비율을 파이차트로 나타내시오.

조건

- 1등석, 2등석, 3등석 구분
- 생존자만을 기준으로 비율 계산

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic = pd.read_csv("./titanic.csv")

survivied =  titanic[titanic["Survived"]==1]
pclass_ratio =  survivied["Pclass"].value_counts().sort_index()
#print(pclass_survivied)
print(pclass_ratio)

plt.pie(
    pclass_ratio,
    labels=["1st","2nd","3rd"],
    autopct="%0.1f%%",
    startangle=90,
    shadow=True,
    explode=(0.06,0,0),
  )

#plt.tight_layout()
plt.title(f"class별 생존자 비율 표시")

```
