### 30. 딕셔너리 {"kim":90, "lee":85, "park":95} 에서 값이 가장 큰 이름을 구하세요.

```python
students = {"kim":90, "lee":85, "park":95}
maxName = None
maxScore = 0
for item in students :
  if students[item] > maxScore :
    maxScore=students[item]
    maxName = item
print(maxName,"===",students.get(maxName))    
```
