### 4. 리스트 [3,7,2,9,4]에서 최댓값을 구하세요.

```python
nums= [3,7,2,9,4]
maxNum = nums[0]
for num in nums :
  if num > maxNum :
    maxNum = num
print(maxNum)  
```

### 4. 리스트 [3,7,2,9,4]에서 최소값을 구하세요.
