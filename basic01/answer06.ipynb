### 1) Student 클래스 만들기

학생 이름(name)과 나이(age)를 속성으로 가지고, <br>
학생 정보를 출력하는 info() 메서드를 가진 클래스를 작성하시오.

```python
class Student :
  def __init__(self,name = "noname",age = 0):
    self.name=name
    self.age=age
  def info(self):
    return f"학생이름 : {self.name}, 나이 : {self.age}"

s = Student("홍길동",20)
print(s.info())

s = Student()
print(s.info())
```

### 2) Rectangle 클래스


가로(width), 세로(height)를 받아 <br>
넓이와 둘레를 계산하는 메서드를 포함한 Rectangle 클래스를 만드시오.

```python
class Rectangle:
  def __init__(self,width,height):
    self.width=width
    self.height=height

  def area(self):
    return self.width*self.height
  
  def perimeter(self):
    return self.width*2+self.height*2
  
r= Rectangle(10,10)
print(r.area())
print(r.perimeter())
```

### 3) Account 은행 계좌 클래스

계좌번호(account_num) <br>
잔액(balance) 

입금, 출금, 잔액 조회 기능을 구현하시오.<br>
잔액 부족 시 출금 불가 메시지 출력.



```python
class Account:
  def __init__(self,account_num,balance=0):
    self.account_num = account_num
    self.balance = balance

  def deposit(self,amount):
    self.balance+=amount
  
  def withdraw(self,amount):
    if self.balance > amount :
      self.balance-=amount
    else :
      print("잔액이 부족합니다.")

  def show_balance(self):
    return self.balance

my_account = Account("001",100000)
my_account.deposit(10000)
print(my_account.show_balance())
my_account.withdraw(30000)
print(my_account.show_balance())
```

### 4) Car 자동차 클래스


모델명(model), 속도(speed)를 갖고 <br>
속도 증가, 감소, 출력 기능을 만들되 속도는 음수가 되지 않도록 하시오.

```python
class Car :
  def __init__(self,model,speed=0):
    self.model= model
    self.speed=speed

  def speed_up(self, amount):
    self.speed+=amount
  
  def speed_down(self, amount):
    self.speed-=amount
  
  def show_speed(self):
    return self.speed
my_car = Car("bentz s class")
my_car.speed_up(100)
print(my_car.show_speed())
my_car.speed_down(50)
print(my_car.show_speed())

  
```

### 5) Calculator 클래스

두 수를 받아 덧셈, 뺄셈, 곱셈, 나눗셈 기능 구현. <br>
0으로 나누면 오류 메시지를 출력.

```python
class Calculator:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  
  def add(self):
    return self.a+self.b
  
  def sub(self):
    return self.a-self.b
  
  def multiple(self):
    return self.a*self.b
  
  def divide(self):
    if self.b == 0:
      return "0으로 나눌 수 없습니다."
    return self.a / self.b
  
my_calc = Calculator(10,5)
print(my_calc.add())
print(my_calc.sub())
print(my_calc.multiple())
print(my_calc.divide())  
```

### 6) Dog 클래스 – 클래스 변수

Dog 클래스를 만들고 객체가 생성될 때마다 클래스 변수 count가 1씩 증가하도록 하시오.

```python
class Dog:
  count=0 # class 내부에 선언된 변수
  # 자바에 있는 접근 제어자가 없음 private
  def __init__(self,name="개"):
    self.__name=name  #__name은 접근이 안됨 즉 private 변수가 된다.
    Dog.count+=1

  def get_name(self):
    return self.__name
  
  def set_name(self, name):
    self.__name=name

dog01 = Dog("해피")
dog02 = Dog("메리")
print(dog01.get_name())
```

```python
class Cat:
  count=0
  def __init__(self,name="고냥이"):
    self.__name=name  #__name은 접근이 안됨 즉 private 변수가 된다.
    Cat.count+=1

  @property # @property를 쓰면 getter가 됨
  def name(self):
    return self.__name
  
  @name.setter
  def name(self, name):
    self.__name=name

my_cat= Cat("너구리")
print(my_cat.name)
my_cat.name="짬타이거"
print(my_cat.name)
```

### 7) Person → Student 상속

Person 클래스: name, greeting() <br>
Student 클래스: major 추가, greeting() 오버라이드

```python
class Person:
  def __init__(self, name="noname"):
    self.name=name
  def greeting(self):
    return f"안녕하세요. 저는 {self.name}입니다."

class Student(Person):
  def __init__(self, name="noname",major="python"):
    super().__init__(name)
    self.major= major
  def greeting(self):
    return f"안녕하세요. 저는 {self.major}전공하는 {self.name}입니다."
  
s= Student("홍길동","컴공")
print(s.greeting())
```
