### 1) Student 클래스 만들기

학생 이름(name)과 나이(age)를 속성으로 가지고, <br>
학생 정보를 출력하는 info() 메서드를 가진 클래스를 작성하시오.

````python
class Student :
  def __init__(self,name = "noname",age = 0):
    self.name=name
    self.age=age
  def info(self):
    return f"학생이름 : {self.name}, 나이 : {self.age}"

s = Student("홍길동",20)
print(s.info())

s = Student()
print(s.info())
````

### 2) Rectangle 클래스


가로(width), 세로(height)를 받아 <br>
넓이와 둘레를 계산하는 메서드를 포함한 Rectangle 클래스를 만드시오.

````python
class Rectangle:
  def __init__(self,width,height):
    self.width=width
    self.height=height

  def area(self):
    return self.width*self.height
  
  def perimeter(self):
    return self.width*2+self.height*2
  
r= Rectangle(10,10)
print(r.area())
print(r.perimeter())
````

### 3) Account 은행 계좌 클래스

계좌번호(account_num) <br>
잔액(balance) 

입금, 출금, 잔액 조회 기능을 구현하시오.<br>
잔액 부족 시 출금 불가 메시지 출력.



````python
class Account:
  def __init__(self,account_num,balance=0):
    self.account_num = account_num
    self.balance = balance

  def deposit(self,amount):
    self.balance+=amount
  
  def withdraw(self,amount):
    if self.balance > amount :
      self.balance-=amount
    else :
      print("잔액이 부족합니다.")

  def show_balance(self):
    return self.balance

my_account = Account("001",100000)
my_account.deposit(10000)
print(my_account.show_balance())
my_account.withdraw(30000)
print(my_account.show_balance())
````

### 5) Calculator 클래스

두 수를 받아 덧셈, 뺄셈, 곱셈, 나눗셈 기능 구현. <br>
0으로 나누면 오류 메시지를 출력.

````python
class Calculator:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  
  def add(self):
    return self.a+self.b
  
  def sub(self):
    return self.a-self.b
  
  def multiple(self):
    return self.a*self.b
  
  def divide(self):
    if self.b == 0:
      return "0으로 나눌 수 없습니다."
    return self.a / self.b
  
my_calc = Calculator(10,5)
print(my_calc.add())
print(my_calc.sub())
print(my_calc.multiple())
print(my_calc.divide())  
````

### 6) Dog 클래스 – 클래스 변수

Dog 클래스를 만들고 객체가 생성될 때마다 클래스 변수 count가 1씩 증가하도록 하시오.

````python
class Dog:
  count=0 # class 내부에 선언된 변수
  # 자바에 있는 접근 제어자가 없음 private
  def __init__(self,name="개"):
    self.__name=name  #__name은 접근이 안됨 즉 private 변수가 된다.
    Dog.count+=1

  def get_name(self):
    return self.__name
  
  def set_name(self, name):
    self.__name=name

dog01 = Dog("해피")
dog02 = Dog("메리")
print(dog01.get_name())
````

````python
class Cat:
  count=0
  def __init__(self,name="고냥이"):
    self.__name=name  #__name은 접근이 안됨 즉 private 변수가 된다.
    Cat.count+=1

  @property # @property를 쓰면 getter가 됨
  def name(self):
    return self.__name
  
  @name.setter
  def name(self, name):
    self.__name=name

my_cat= Cat("너구리")
print(my_cat.name)
my_cat.name="짬타이거"
print(my_cat.name)
````

### 7) Person → Student 상속

Person 클래스: name, greeting() <br>
Student 클래스: major 추가, greeting() 오버라이드

````python
class Person:
  def __init__(self, name="noname"):
    self.name=name
  def greeting(self):
    return f"안녕하세요. 저는 {self.name}입니다."

class Student(Person):
  def __init__(self, name="noname",major="python"):
    super().__init__(name)
    self.major= major
  def greeting(self):
    return f"안녕하세요. 저는 {self.major}전공하는 {self.name}입니다."
  
s= Student("홍길동","컴공")
print(s.greeting())
````

## **8) 문제 8 – 클래스 메서드 / 정적 메서드**

MathUtil 클래스 만들기:

- 클래스 변수: PI = 3.14159
- 정적 메서드: add(a, b)
- 클래스 메서드: circle_area(cls, r) → PI * r^2

````python
class MathUtil:
  PI = 3.141592

  @staticmethod
  def add(a,b):
    return a+b
  
  @classmethod
  def circle_area(cls, r) :
    return cls.PI*(r**2)

print(MathUtil.add(10,10))
print(MathUtil.circle_area(4))
````

### 9) 문제 9 – 다형성(Polymorphism)

Animal 클래스 → speak() 만들어 "동물 소리" 출력 <br>
Dog, Cat 클래스를 만들어 speak()를 각각 다르게 오버라이드:

````python
class Animal:
  def speak(self):
    return "우우웅"

class Dog(Animal):
  def speak(self):
    return "멍멍"

class Cat(Animal):
  def speak(self):
    return "냐옹"
animals = [Dog(),Cat(),Animal()]
for animal in animals:
  print(animal.speak())
````

### 11) 온도 변환 클래스 – property 활용

**요구사항**

- `Temperature` 클래스를 만든다. 
- 내부(private) 변수로 섭씨 온도 `__celsius`를 가진다.
- 생성자:
    - `celsius` 값을 받아 저장 (기본값 0)
- `celsius` 프로퍼티:
    - getter: 현재 섭씨 온도 반환
    - setter: 섭씨 온도가 -273.15도(절대영도)보다 작아지지 않도록 체크
        
        (그보다 작은 값이 들어오면 예외 발생 또는 무시 처리)
        
- `fahrenheit` 프로퍼티:
    - getter: 현재 섭씨 값을 화씨로 변환해서 반환
        
        (공식: `F = C * 9/5 + 32`)
        
    - setter: 화씨 값이 주어졌을 때 이를 섭씨로 변환해서
        
        내부의 `__celsius`에 저장
        
        (공식: `C = (F - 32) * 5/9`)
        
- 사용 예시(의도):

```python
t = Temperature(25)
print(t.celsius)      # 25
print(t.fahrenheit)   # 77.0

t.fahrenheit = 212
print(t.celsius)      # 100.0

t.celsius = -300      # 오류 또는 무시되도록 처리

```

````python
class Temperature:
  def __init__(self,celsius=0):
    self.__celsius= celsius
  
  @property
  def celsius(self):
    return self.__celsius
  
  @celsius.setter
  def celsius(self,value):
    if value < -273:
      raise ValueError("섭씨 온도는 -273아래로 쓸 수 없습니다.")
    self.__celsius=value

  @property
  def fahrenheit(self):
    return self.__celsius* 9/5 + 32

  @fahrenheit.setter
  def fahrenheit(self,value):
    c = (value - 32) * 5/9
    if c <- 173 :
      raise ValueError("섭씨 온도는 -273아래로 쓸 수 없습니다.")
    self.__celsius=c

t = Temperature(-3)
print(t.celsius)
print(t.fahrenheit)
````
