### 1) 리스트 평균 구하기 (예외 처리 포함)

정수로 이루어진 리스트 `nums`를 받아 **평균값을 반환하는 함수** `calc_avg(nums)`를 작성하세요.

단, 리스트가 비어 있으면 `None`을 반환하세요.

```python
# 예시
print(calc_avg([10, 20, 30]))  # 20.0
print(calc_avg([]))            # None

```

````python
def calc_avg(nums):
  if not nums :
      return None
  return sum(nums) / len(nums)
print(calc_avg([10, 20, 30,36]))  # 20.0
print(calc_avg([]))            # None
````

### 2) 문자열 리스트에서 특정 길이 이상만 필터링

문자열 리스트 `words`와 정수 `n`을 받아,

길이가 `n` 이상인 단어만 모아서 **새 리스트로 반환**하는 함수 `filter_by_length(words, n)`를 작성하세요.

```python
# 예시
print(filter_by_length(["hi", "hello", "python", "a"], 3))
# ["hello", "python"]

```

````python
def filter_by_length(words, num):
  result = []
  for word in words :
    if len(word) >= num :
      result.append(word)
  return result
print(filter_by_length(["hi", "hello", "python", "a"], 3))
````

### 3) 가변 인자를 이용한 합계·평균 함수

숫자를 여러 개 받을 수 있는 함수 `stats(*nums)`를 작성하세요.<br>
반환값은 `(합계, 평균)` 형태의 **튜플**이어야 합니다.<br>
숫자가 하나도 안 들어오면 `(0, 0)`을 반환하세요.<br>
```python
# 예시
print(stats(1, 2, 3, 4))  # (10, 2.5)
print(stats())            # (0, 0)

```

````python
def stats(*nums) :
  if not nums :
    return 0,0
  total = sum(nums)
  avg =  total/len(nums)
  return total, avg
print(stats(1, 2, 3, 4))  # (10, 2.5)
print(stats())            # (0, 0)
````

### 4) 딕셔너리에서 특정 값 이상인 항목만 추출

이름-점수 딕셔너리 `scores`와 `cutline`을 받아,<br>
점수가 `cutline` 이상인 항목만 모아 **새 딕셔너리**로 반환하는 함수<br>
`filter_pass(scores, cutline)`를 작성하세요.

```python
# 예시
scores = {"홍길동": 80, "김유신": 60, "이순신": 95}
print(filter_pass(scores, 70))
# {"홍길동": 80, "이순신": 95}

```

````python
def filter_pass(scores, cutline) :
  result = {}
  for name, score in scores.items():
    if score >= cutline:
      result[name] = score
  return result
scores = {"홍길동": 80, "김유신": 60, "이순신": 95}
print(filter_pass(scores, 70))
````

### 5) 재귀 함수를 이용한 리스트 합계

정수 리스트 `nums`의 합을 **재귀 함수**로 구하는 `recursive_sum(nums)` 함수를 작성하세요.<br>
반복문(`for`, `while`)은 사용하지 마세요.

```python
# 예시
print(recursive_sum([1, 2, 3, 4]))  # 10
print(recursive_sum([]))            # 0

```

````python
def recursive_sum(nums):
  if not nums:
    return 0
  return nums[0]+recursive_sum(nums[1:])
print(recursive_sum([1, 2, 3, 4]))  # 10
print(recursive_sum([]))            # 0

def factorial(num):
  if num==0 or num==1:
    return 1
  return num*factorial(num-1)
print(factorial(20))
````

### 6) 문자열에서 숫자만 추출하여 합 구하기

문자열 `s`를 받아, 그 안에 포함된 모든 숫자 문자(0~9)를 찾아<br>
정수로 변환한 뒤 **합계를 반환**하는 함수 `sum_digits(s)`를 작성하세요.

```python
# 예시
print(sum_digits("a1b2c3"))   # 6
print(sum_digits("abc"))      # 0
print(sum_digits("2025year")) # 2+0+2+5 = 9

```

````python
def sum_digits(str):
  total = 0
  for ch in str:
    if ch.isdigit():
      total+=int(ch)
  return total
print(sum_digits("a1b2c3"))   # 6
print(sum_digits("abc"))      # 0
print(sum_digits("2025year")) # 2+0+2+5 = 9
````

### 7) 이메일 형식 간단 검증 함수

문자열 `email`을 인자로 받아, <br>
다음 조건을 모두 만족하면 `True`, 아니면 `False`를 반환하는 함수 `is_valid_email(email)`을 작성하세요.

- `@`가 1개 이상 포함
- `.` 이 `@` 뒤에 최소 1개 이상 있음

```python
# 예시
print(is_valid_email("test@example.com"))  # True
print(is_valid_email("testexample.com"))   # False
print(is_valid_email("test@examplecom"))   # False

```

````python
def is_valid_email(email):
  if "@" not in email:
    return False
  at_idx = email.find("@")
  if "." in email[at_idx+1:]:
    return True
  return False

print(is_valid_email("test@example.com"))  # True
print(is_valid_email("testexample.com"))   # False
print(is_valid_email("test@examplecom"))   # False
````

### 8) 2차원 리스트(행렬)의 전치(transpose)

2차원 리스트(행렬) `matrix`를 받아, **행과 열을 뒤집은 전치 행렬**을 <br>
새 2차원 리스트로 반환하는 함수 `transpose(matrix)`를 작성하세요.<br>

(반복문으로 직접 구현, zip 사용 금지)

```python
# 예시
m = [
    [1, 2, 3],
    [4, 5, 6]
]
print(transpose(m))
# [
#   [1, 4],
#   [2, 5],
#   [3, 6]
# ]

```

````python
def transpose(matrix):
  if not matrix:
    return []
  
  rows = len(matrix)
  cols = len(matrix[0])
  
  result = []
  for col in range(cols):
    new_row =[]
    for row in range(rows):
      new_row.append(matrix[row][col])
    result.append(new_row)
  return result
m = [
    [1, 2, 3],
    [4, 5, 6]
]
print(transpose(m))
````

### 9) 키워드 인자를 받아 출력 형식 만들기

키워드 인자를 가변으로 받는 함수 `format_user(**info)`를 작성하세요.<br>
다음과 같이 호출될 수 있으며,<br>
`"name: 홍길동, age: 30"` 형식의 문자열을 반환해야 합니다.<br>
키의 순서는 `name`, `age`, `email` 순서를 우선으로 하고,<br>
그 외 키는 뒤에 붙여도 됩니다.

```python
# 예시
print(format_user(name="홍길동", age=30))
# "name: 홍길동, age: 30"

print(format_user(name="Tom", age=25, email="tom@test.com"))
# "name: Tom, age: 25, email: tom@test.com"

```

````python
def format_user(**info):
  result = []
  if "name" in info:
    result.append(f"name:{info['name']}")
  if "age" in info:
    result.append(f"age:{info['age']}")
  if "email" in info:
    result.append(f"email:{info['email']}")
  for key,value in info.items():
    if key not in ("name","age","email"): 
      result.append(f"{key}:{value}")
  return ",".join(result)

print(format_user(address="일산",email="jkfdkjfs@fdsjfkd.com",age=30,name="홍길동"))
  
````

### 10) 함수형 스타일의 필터 함수 만들기

리스트 `nums`와 **판별 함수** `pred`를 인자로 받아,<br>
`pred(x)`가 `True`인 값들만 모아 새 리스트로 반환하는 함수 `my_filter(nums, pred)`를 작성하세요.

```python
# 예시
def is_even(x):
    return x % 2 == 0

print(my_filter([1,2,3,4,5,6], is_even))
# [2, 4, 6]

```

````python
def my_filter(nums,predicate):
  result = []
  for num in nums:
    if predicate(num):
      result.append(num)
  return result

def is_even(x):
    return x % 2 == 0

print(my_filter([1,2,3,4,5,6], is_even))
````

```python
nums = [1, 2, 3, 4]

```

### 12) 짝수만 필터링 (filter + lambda)

```python
nums = [1, 2, 3, 4, 5, 6]
```

### 13) 문자열 길이 기준 정렬 (sorted + lambda)

문자열 리스트 `words`를 <br>
길이가 짧은 순서대로 **오름차순 정렬**하세요.

```python
words = ["python", "is", "fantastic", "good"]
# 결과 예: ["is", "good", "python", "fantastic"]

```

````python
words = ["python", "is", "fantastic", "good"]
sorted_words = sorted(words,key=lambda word: len(word), reverse=True)
print(sorted_words)
````

### 15) 딕셔너리 리스트를 age 기준 내림차순 정렬

다음과 같은 리스트가 있을 때, `age` 키를 기준으로 **내림차순 정렬**하세요.
