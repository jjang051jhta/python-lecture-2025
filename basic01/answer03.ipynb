## **1. 리스트 중복 제거 + 정렬**

정수 리스트가 주어졌을 때,

중복을 제거하고 오름차순 정렬한 새로운 리스트를 직접 구현하여 출력하시오.

(`set()`만 단독 사용 금지)

````python
nums = [1,3,4,5,3,4,2,1,3,3,4,6,7,8,9,2,7,6,5,8,8]
unique = []
for num in nums:
  if  num not in unique:
    unique.append(num)
unique.sort()
print(unique)
  
````

````python
nums = [1,3,4,5,3,4,2,1,3,3,4,6,7,8,9,2,7,6,5,8,8]
unique = sorted(set(nums))
print(unique)
````

## **2. 두 리스트로 딕셔너리 만들기**

다음 두 리스트가 있다
```python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
```
두 리스트를 이용해 `{"홍길동":90, ...}` 형태의 딕셔너리를 만들고,
평균 점수보다 높은 학생만 출력하시오.

````python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
students = {}
for i in range(len(names)):
  students[names[i]] = scores[i]
print(students)

avg = sum(scores) / len(scores)
print(avg)

for name in students:
  if students[name] > avg:
    print(name,":", students[name])
````

````python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
students = dict(zip(names,scores))
print(students)
avg = sum(scores) / len(scores)
print(avg)
for name,score in students.items():
  if score > avg :
    print(name,":",score)
````

````python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
students = {name:score for name,score in zip(names,scores)}
print(students)
avg = sum(scores) / len(scores)
print(avg)
for name,score in students.items():
  if score > avg :
    print(name,":",score)
````

## **3. 딕셔너리에서 점수별 등급 나누기**

다음 딕셔너리가 있다:
```python
students = {"홍길동":92, "김유신":76, "이순신":88, "강감찬":59}
```
점수 기준에 따라 A~F 등급을 부여하여<br>
이름 - 점수 - 등급` 형식으로 출력하시오.

````python
students = {"홍길동":92, "김유신":76, "이순신":88, "강감찬":59}
for name,score in students.items():
  if score>=90: 
    grade="A"
  elif score>=80:
    grade="B"
  elif score>=70:
    grade="C"
  elif score>=60:
    grade="D"
  else :
    grade="F"
  print(name,":",score,":",grade)
  
````

## **4. 리스트 안 딕셔너리에서 최고 점수 찾기**

다음 리스트에서 최고 점수를 받은 학생을 찾으시오
```python
data = [
    {"name": "홍길동", "score": 90},
    {"name": "김유신", "score": 80},
    {"name": "이순신", "score": 95}
]
```


````python
data = [
    {"name": "홍길동", "score": 90},
    {"name": "김유신", "score": 80},
    {"name": "이순신", "score": 95}
]
maxScore = 0
maxName = ""
for s in data:
  if s["score"] > maxScore:
    maxScore = s["score"]
    maxName = s["name"]
print("최고점수 : ",maxScore, maxName)
````

## 5. 단어 빈도수 세기
```python
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
```
각 단어가 몇 번 등장했는지 딕셔너리에 저장하고 출력하시오.

````python
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
frequency = {}
for word in words:
  if word in frequency:
    frequency[word]+=1
  else:
    frequency[word]=1
print(frequency)
````
