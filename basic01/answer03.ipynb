## **1. 리스트 중복 제거 + 정렬**

정수 리스트가 주어졌을 때,

중복을 제거하고 오름차순 정렬한 새로운 리스트를 직접 구현하여 출력하시오.

(`set()`만 단독 사용 금지)

````python
nums = [1,3,4,5,3,4,2,1,3,3,4,6,7,8,9,2,7,6,5,8,8]
unique = []
for num in nums:
  if  num not in unique:
    unique.append(num)
unique.sort()
print(unique)
  
````

````python
nums = [1,3,4,5,3,4,2,1,3,3,4,6,7,8,9,2,7,6,5,8,8]
unique = sorted(set(nums))
print(unique)
````

## **2. 두 리스트로 딕셔너리 만들기**

다음 두 리스트가 있다
```python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
```
두 리스트를 이용해 `{"홍길동":90, ...}` 형태의 딕셔너리를 만들고,
평균 점수보다 높은 학생만 출력하시오.

````python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
students = {}
for i in range(len(names)):
  students[names[i]] = scores[i]
print(students)

avg = sum(scores) / len(scores)
print(avg)

for name in students:
  if students[name] > avg:
    print(name,":", students[name])
````

````python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
students = dict(zip(names,scores))
print(students)
avg = sum(scores) / len(scores)
print(avg)
for name,score in students.items():
  if score > avg :
    print(name,":",score)
````

````python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
students = {name:score for name,score in zip(names,scores)}
print(students)
avg = sum(scores) / len(scores)
print(avg)
for name,score in students.items():
  if score > avg :
    print(name,":",score)
````

## **3. 딕셔너리에서 점수별 등급 나누기**

다음 딕셔너리가 있다:
```python
students = {"홍길동":92, "김유신":76, "이순신":88, "강감찬":59}
```
점수 기준에 따라 A~F 등급을 부여하여<br>
이름 - 점수 - 등급` 형식으로 출력하시오.

````python
students = {"홍길동":92, "김유신":76, "이순신":88, "강감찬":59}
for name,score in students.items():
  if score>=90: 
    grade="A"
  elif score>=80:
    grade="B"
  elif score>=70:
    grade="C"
  elif score>=60:
    grade="D"
  else :
    grade="F"
  print(name,":",score,":",grade)
  
````

## **4. 리스트 안 딕셔너리에서 최고 점수 찾기**

다음 리스트에서 최고 점수를 받은 학생을 찾으시오
```python
data = [
    {"name": "홍길동", "score": 90},
    {"name": "김유신", "score": 80},
    {"name": "이순신", "score": 95}
]
```


````python
data = [
    {"name": "홍길동", "score": 90},
    {"name": "김유신", "score": 80},
    {"name": "이순신", "score": 95}
]
maxScore = 0
maxName = ""
for s in data:
  if s["score"] > maxScore:
    maxScore = s["score"]
    maxName = s["name"]
print("최고점수 : ",maxScore, maxName)
````

## 5. 단어 빈도수 세기
```python
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
```
각 단어가 몇 번 등장했는지 딕셔너리에 저장하고 출력하시오.

````python
words = ["apple", "banana", "apple", "orange", "banana", "apple"]
frequency = {}
for word in words:
  if word in frequency:
    frequency[word]+=1
  else:
    frequency[word]=1
print(frequency)
````

## 6. 튜플 리스트 정렬하기
```python
items = [("apple", 3), ("banana", 1), ("orange", 5)]
```
두 번째 값(숫자) 기준으로 오름차순 정렬한 새로운 리스트를 만들고 출력하시오.

````python
items = [("apple",3),("banana",1),("orange",5)]
sorted_items = items[:]
for i in range(len(sorted_items)):
    for j in range(i+1,len(sorted_items)):
        if sorted_items[i][1] > sorted_items[j][1]:
            sorted_items[i], sorted_items[j] = sorted_items[j], sorted_items[i]
print(sorted_items)
````

## 7. 조건에 맞는 딕셔너리 삭제
```python
prices = {"item1":1200, "item2":500, "item3":3000, "item4":800}
```
1000 미만의 상품을 딕셔너리에서 제거한 뒤 출력하시오.

````python
prices = {"item1":1200, "item2":500, "item3":3000, "item4":800}
deleteItems = []
for item,price in prices.items():
  print(item,"==",price)
  if price<1000:
    deleteItems.append(item)
print(deleteItems)
for item in deleteItems:
  del prices[item]
print(prices)
````

## 8. 2차원 리스트 행별 합 + 전체 합
```python
matrix = [
    [1,2,3],
    [4,5,6],
    [7,8,9]
]
```
- 각 행의 합
- 전체 합  
을 출력하시오.

````python
matrix = [[1,2,3],[4,5,6],[7,8,9]]
rowSum=[]
total =0
for row in matrix:
  #print(row)
  sum=0
  for r in row:
    sum+=r
  rowSum.append(sum)
  total+=sum
print("행의 합 : ",rowSum)
print("전체 합 : ",total)
````

## 9. 합격/불합격 분리
```python
scores = [45, 67, 89, 72, 50, 38, 91]
```
60점 이상은 합격, 미만은 불합격으로 분류하여

두 개의 리스트에 담아 출력하시오.

````python
scores = [45, 67, 89, 72, 50, 38, 91]
passed = []
failed = []
for score in scores:
  if score>60 :
    passed.append(score)
  else :
    failed.append(score)
print("합격 : ",passed)
print("불합격 : ",failed)
````

## 10. 연속된 값 압축하기
```python
data = [1,1,2,2,3,3,3,2,2,1]
```
연속된 값은 하나만 남기고 압축한 리스트를 만드시오.
예: `[1,2,3,2,1]`


````python
data = [1,1,2,2,3,3,3,2,2,1]
compressed = []
prev = None
for num in data:
  if prev is None or num !=prev:
    compressed.append(num)
    prev  = num
print(compressed) 
````

## 11. 리스트에서 소수만 골라내기
```python
nums = [2,3,4,5,10,11,17]
```
이 중 **소수만 리스트로 만들어 출력**하시오.
(소수 판별 직접 구현)

````python
nums = [2,3,4,5,10,11,17]
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
primes = []
for n in nums:
    if is_prime(n):
        primes.append(n)
#print(is_prime(488540985890435541))
print(primes)
````

## 12. 회문만 골라내기
```python
words = ["level", "apple", "noon", "madam", "python"]
```
회문(palindrome)인 단어만 새로운 리스트에 담아 출력하시오.

````python
words = ["level", "apple", "noon", "madam", "python"]
palindrome = []
for word in words :
  if word==word[::-1]:
    palindrome.append(word)
print(palindrome)
````

## 13. 딕셔너리 뒤집기 (key ↔ value)
```python
data = {"a":1, "b":2, "c":3}
```
`{1:"a", 2:"b", 3:"c"}` 형태의 새로운 딕셔너리를 만들어 출력하시오.

````python
data = {"a":1, "b":2, "c":3}
reversed = {}
for k,v in data.items():
  reversed[v] = k
print(reversed)
print(reversed[1])
````

## **14. 문자열에서 가장 많이 등장한 문자 찾기**

공백 제외한 문자열이 주어졌을 때, <br>
가장 많이 등장한 문자를 찾아 출력하시오.<br>
여러 개여도 아무거나 하나만 출력해도 됨.

````python
text = "hello python life is short you need python"
freq = {}
for ch in text:
  if ch == " ":
    continue
  if ch in freq:
    freq[ch]+=1
  else :
    freq[ch]=1
print(freq)
maxChar = None
maxCount=-1
for ch,count in freq.items():
  if count>maxCount:
    maxCount=count
    maxChar=ch
print(f"가장 많이 나온 문자 : {maxChar}는 {maxCount}번 나왔습니다.")
````
