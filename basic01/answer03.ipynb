## **1. 리스트 중복 제거 + 정렬**

정수 리스트가 주어졌을 때,

중복을 제거하고 오름차순 정렬한 새로운 리스트를 직접 구현하여 출력하시오.

(`set()`만 단독 사용 금지)

````python
nums = [1,3,4,5,3,4,2,1,3,3,4,6,7,8,9,2,7,6,5,8,8]
unique = []
for num in nums:
  if  num not in unique:
    unique.append(num)
unique.sort()
print(unique)
  
````

````python
nums = [1,3,4,5,3,4,2,1,3,3,4,6,7,8,9,2,7,6,5,8,8]
unique = sorted(set(nums))
print(unique)
````

## **2. 두 리스트로 딕셔너리 만들기**

다음 두 리스트가 있다
```python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
```
두 리스트를 이용해 `{"홍길동":90, ...}` 형태의 딕셔너리를 만들고,
평균 점수보다 높은 학생만 출력하시오.

````python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
students = {}
for i in range(len(names)):
  students[names[i]] = scores[i]
print(students)

avg = sum(scores) / len(scores)
print(avg)

for name in students:
  if students[name] > avg:
    print(name,":", students[name])
````

````python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
students = dict(zip(names,scores))
print(students)
avg = sum(scores) / len(scores)
print(avg)
for name,score in students.items():
  if score > avg :
    print(name,":",score)
````

````python
names = ["홍길동", "김유신", "이순신"]
scores = [90, 80, 95]
students = {name:score for name,score in zip(names,scores)}
print(students)
avg = sum(scores) / len(scores)
print(avg)
for name,score in students.items():
  if score > avg :
    print(name,":",score)
````
