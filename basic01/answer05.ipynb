## 3. 문제: data.txt 파일 숫자들 중 가장 큰 값과 가장 작은 값을 출력하세요.
```python

```

````python

nums = []
with open("data.txt","r",encoding="utf-8") as f:
  for line in f:
    num = int(line.strip())
    nums.append(num)
print(f"max:{max(nums)}")
print(f"min:{min(nums)}")
````

````python
with open("data.txt","r",encoding="utf-8") as f:
  lines = f.readlines()
  nums = [int(line.strip()) for line in lines]  #list comprehension

print(f"max:{max(nums)}")
print(f"min:{min(nums)}")
````

## 4. 문제: data.txt 파일에서 짝수만 골라 even.txt 파일로 저장하세요.

````python
import os

os.makedirs("test", exist_ok=True)

evens=[]
with open("data.txt","r",encoding="utf-8") as f:
  for line in f:
    num =  int(line.strip())
    if num%2 == 0:
      evens.append(str(num)+"\n")

with open("even.txt","w", encoding="utf-8") as f:
  f.writelines(evens)
````


````python
with open("message.txt","r",encoding="utf-8") as src:
  lines = src.readlines()
with open("result.txt","w",encoding="utf-8") as dest:
  for line in lines:
    line = line.strip("\n")
    dest.write(line+"(end)\n")
````

## 9. 문제: sentence.txt 파일에서 가장 많이 등장한 단어를 출력하세요.
```python
apple banana apple orange kiwi banana apple
banana grape kiwi apple strawberry apple orange
```

````python
word_count={}
with open("sentence.txt","r",encoding="utf-8") as f:
  #print(f.read())
  for line in f:
    words = line.split()
    for word in words:
      if word in word_count:
        word_count[word] += 1
      else:
        word_count[word] = 1
print(word_count)
max_word=None
max_count=0
for word, count in word_count.items():
  if count > max_count:
    max_count=count
    max_word=word
print(f"가장 많이 등장한 단어 {max_word} 총 {max_count}번 등장했습니다.")
  

````

````python
from collections import Counter
with open("sentence.txt","r",encoding="utf-8") as f:
  text= f.read()
words = text.split()
print(words)
counter = Counter(words)
print(counter)
print(counter.most_common(1))
max_word,max_count=counter.most_common(1)[0]
print(f"가장 많이 등장한 단어 {max_word} 총 {max_count}번 등장했습니다.")
````

## 10. 문제: sentence.txt 파일에서 특정 단어의 등장 횟수를 출력하세요.

````python
target = input("찾을 단어를 입력하세요 : ").strip()
with open("sentence.txt","r",encoding="utf-8") as f:
  text= f.read()
words = text.split()
#print(words)
#count = sum(1 for word in words if word==target)  sum generator 조건을 만족하면 1씩 증가
count = len(list(filter(lambda word:word == target, words)))
print(f"찾는 단어 {target}은 총 {count}번 등장합니다.")
````

## 11. 문제: students.csv 파일에서 나이가 가장 많은 학생을 출력하세요.
```python
홍길동,22,컴퓨터공학
김유신,25,신소재공학
이순신,21,컴퓨터공학
강감찬,24,전자공학
유관순,23,기계공학
안중근,30,컴퓨터공학
```

````python
oldest_student=None
oldest_age = -1
with open("students.csv","r",encoding="utf-8") as f:
  for line in f:
    name,str_age,major = line.strip().split(",")
    age = int(str_age)
    if age>oldest_age :
      oldest_age=age
      oldest_student= (name,age,major)
print("나이가 가장 많은 학생 : ",oldest_student)
````

````python
with open("students.csv","r",encoding="utf-8") as f:
  students = [line.strip().split(",") for line in f]
print(students)  
students = [(name,int(age), major) for name,age,major in students]
print(students)  
oldest = max(students,key=lambda x:x[1])
print(oldest)
````
