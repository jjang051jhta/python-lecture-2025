## 1. 문제: scores.txt 파일의 평균 점수를 출력하세요.
```python

김유신 45
이순신 97
강감찬 76
유관순 55
안중근 34
```

````python
f = open("scores.txt","r",encoding="utf-8")
total=0
count=0
for line in f:
  name,score = line.strip().split()
  total+=int(score)
  count+=1
avg = total/count
print(avg)
f.close()

with open("scores.txt","r",encoding="utf-8") as f:
  total=0
  count=0
  for line in f:
    name,score = line.strip().split()
    total+=int(score)
    count+=1
  avg = total/count
  print(avg)


````

## 3. 문제: data.txt 파일 숫자들 중 가장 큰 값과 가장 작은 값을 출력하세요.
```python
10
3
55
27
91
42
18
7
60
```

````python

nums = []
with open("data.txt","r",encoding="utf-8") as f:
  for line in f:
    num = int(line.strip())
    nums.append(num)
print(f"max:{max(nums)}")
print(f"min:{min(nums)}")
````

````python
with open("data.txt","r",encoding="utf-8") as f:
  lines = f.readlines()
  nums = [int(line.strip()) for line in lines]  #list comprehension

print(f"max:{max(nums)}")
print(f"min:{min(nums)}")
````

````python
with open("data.txt","r",encoding="utf-8") as f:
  lines = f.readlines()
  nums = list(map(lambda num: int(num.strip()), lines))

print(f"max:{max(nums)}")
print(f"min:{min(nums)}")
````

## 4. 문제: data.txt 파일에서 짝수만 골라 even.txt 파일로 저장하세요.

````python
import os

os.makedirs("test", exist_ok=True)

evens=[]
with open("data.txt","r",encoding="utf-8") as f:
  for line in f:
    num =  int(line.strip())
    if num%2 == 0:
      evens.append(str(num)+"\n")

with open("even.txt","w", encoding="utf-8") as f:
  f.writelines(evens)
````

## 5. 문제: message.txt 파일의 모든 줄 끝에 (end) 를 붙여 result.txt 로 저장하세요.
```python
안녕하세요
반갑습니다
파이썬 파일 입출력 공부 중입니다
열심히 해봅시다
```

````python
with open("message.txt","r",encoding="utf-8") as src:
  lines = src.readlines()
with open("result.txt","w",encoding="utf-8") as dest:
  for line in lines:
    line = line.strip("\n")
    dest.write(line+"(end)\n")
````

## 6. 문제: words.txt 파일에서 모음(a,e,i,o,u)으로 시작하는 단어만 출력하세요.

````python
vowels = "aeiouAEIOU"
with open("words.txt","r",encoding="utf-8") as f:
  for line in f:
    word = line.strip()
    if word and word[0] in vowels:
      print(word)
````

## 7. 문제: 글자 수 5 이상인 단어를 대문자로 바꿔 long_words.txt 에 저장하세요.

````python
with open("words.txt","r",encoding="utf-8") as src, \
     open("long_words.txt","w",encoding="utf-8") as dest:
     for line in src:
          word = line.strip()
          if len(word) > 5 :
               dest.write(word.upper()+"\n")
````

## 8. 문제: numbers.txt 에 있는 숫자들의 합과 평균을 구하세요.

## 9. 문제: sentence.txt 파일에서 가장 많이 등장한 단어를 출력하세요.
```python
apple banana apple orange kiwi banana apple
banana grape kiwi apple strawberry apple orange
```

````python
word_count={}
with open("sentence.txt","r",encoding="utf-8") as f:
  #print(f.read())
  for line in f:
    words = line.split()
    for word in words:
      if word in word_count:
        word_count[word] += 1
      else:
        word_count[word] = 1
print(word_count)
max_word=None
max_count=0
for word, count in word_count.items():
  if count > max_count:
    max_count=count
    max_word=word
print(f"가장 많이 등장한 단어 {max_word} 총 {max_count}번 등장했습니다.")
  

````
