## 1. 문제: scores.txt 파일의 평균 점수를 출력하세요.
```python
홍길동 82
김유신 45
이순신 97
강감찬 76
유관순 55
안중근 34
```

````python
f = open("scores.txt","r",encoding="utf-8")
total=0
count=0
for line in f:
  name,score = line.strip().split()
  total+=int(score)
  count+=1
avg = total/count
print(avg)
f.close()

with open("scores.txt","r",encoding="utf-8") as f:
  total=0
  count=0
  for line in f:
    name,score = line.strip().split()
    total+=int(score)
    count+=1
  avg = total/count
  print(avg)


````

## 2. 문제: scores.txt 파일에서 50점 미만 학생 이름만 출력하세요.

````python
with open("scores.txt","r",encoding="utf-8") as f:
  for line in f:
    name,score  = line.strip().split()
    if int(score)<50:
      print(name)
````

## 3. 문제: data.txt 파일 숫자들 중 가장 큰 값과 가장 작은 값을 출력하세요.

````python
nums = []
with open("data.txt","r",encoding="utf-8") as f:
  for line in f:
    num = int(line.strip())
    nums.append(num)
print(f"max:{max(nums)}")
print(f"min:{min(nums)}")
````

````python
with open("data.txt","r",encoding="utf-8") as f:
  lines = f.readlines()
  nums = [int(line.strip()) for line in lines]  #list comprehension

print(f"max:{max(nums)}")
print(f"min:{min(nums)}")
````

````python
with open("data.txt","r",encoding="utf-8") as f:
  lines = f.readlines()
  nums = list(map(lambda num: int(num.strip()), lines))

print(f"max:{max(nums)}")
print(f"min:{min(nums)}")
````
