### 2) `"2025-12-10"` 문자열을 datetime 객체로 변환한 뒤, 요일 이름(예: "Wednesday")을 출력하시오.

```python
from datetime import datetime
str_now = "2025-12-10"
now = datetime.strptime(str_now,"%Y-%m-%d")
weekday = now.strftime("%A")
kor_weekdays = ["월요일","화요일","수요일","목요일","금요일","토요일","일요일",]
print(f"{now}의 요일은 {weekday}")
print(f"{now}의 한글 요일은 {kor_weekdays[now.weekday()]}")
```

### 4) `"2025-12-10"` 기준으로 1주일 후, 1개월 후, 1년 후 날짜를 각각 구하시오.

```python
from datetime import datetime, date, timedelta
base_date =  date(2025,12,10)
after_one_week = base_date+timedelta(weeks=1)
temp_month = base_date.month+1
temp_year = base_date.year
if temp_month > 12:
  temp_month = 1
  temp_year+=1
temp_day = base_date.day
while True:
  try:
    after_one_month = date(temp_year,temp_month,temp_day)
    break
  except ValueError:
    temp_day-=1

try:
  after_one_year =  base_date.replace(year=base_date.year+1)
except ValueError:
  after_one_year =  base_date.replace(year=base_date.year+1, day=28)

print(f"기준 날짜 : {base_date}")
print(f"1주일 후 : {after_one_week}")
print(f"1달 후 : {after_one_month}")
print(f"1년 후 : {after_one_year}")
```

### 5) 2025년 12월 한 달 동안의 모든 월요일 날짜를 리스트로 출력하시오.

```python
from datetime import date, timedelta

year = 2025
month = 12
base_date =  date(year, month,1)
mondays = []

while base_date.month == month:
  if base_date.weekday()==0:
    mondays.append(base_date)
  base_date+= timedelta(days=1)

for monday in mondays:
  print(monday)
```

### 6) 두 날짜 `"2025-12-10"`과 `"2024-05-03"` 의 차이를

`n일 (n주 m일)` 형태로 출력하시오.

```python
from datetime import date
d01 = date(2025,12,10)
d02 = date(2024,5,3)

diff_days = (d01-d02).days

#weeks = diff_days//7
#days = diff_days % 7
weeks,days = divmod(diff_days,7)
print(f"날짜 차이 : {diff_days}일 = {weeks}주 {days}일")
```

```python
from datetime import date
from dateutil.relativedelta import relativedelta

d01 = date(2025,12,10)
d02 = date(2024,5,3)

diff_days = relativedelta(d01,d02)
print(f"날짜 차이 : {diff_days.years}년 {diff_days.months}개월 {diff_days.days}일")
```

### 문제 8

"2025/12/10 15:30" 을 datetime 객체로 변환하고 이를 "2025-12-10 오후 03:30" 형태로 다시 출력하시오.


```python
from datetime import date
from dateutil.relativedelta import relativedelta

str_date = "2025/12/10 15:30"
formated_date = datetime.strptime(str_date,"%Y/%m/%d %H:%M")
hour_24 = formated_date.hour
apm =  "오전" if hour_24 < 12 else "오후"
hour_12 = hour_24%12
if hour_12==0:
  hour_12=12

print(f"{formated_date.strftime('%Y-%m-%d')} {apm} {hour_12:02d}:{formated_date.minute:02d}")
```

### 9) 특정 datetime 객체에서 ‘연도만 변경’하여 2년 뒤 날짜를 만드는 코드를 작성하시오.

```python
from datetime import date
from dateutil.relativedelta import relativedelta

now = datetime.now()
print("현재 : ",now)
try:
  two_add_year = now.replace(year=now.year+2)
except ValueError :
  two_add_year = now.replace(year=now.year+2, day=28)
print("2년뒤 : ",two_add_year)

two_add_year02 = now+relativedelta(years=+2)
print("2년뒤 : ",two_add_year02)


```

### 10) 오늘 날짜가 속한 주의 월~일까지 날짜를 리스트로 만드시오.

### 11) 두 시간 `"09:30"` 과 `"18:30"` 의 차이가 12시간 이상이면 `"장시간 근무"`, 아니면 `"정상 근무"` 를 출력하시오.

```python
from datetime import date
from dateutil.relativedelta import relativedelta

str_start_time = "09:30"
str_end_time = "23:30"
format =  "%H:%S"
start_time = datetime.strptime(str_start_time,format)
end_time = datetime.strptime(str_end_time,format)
diff = end_time - start_time
# print(type(diff))
# print(diff.total_seconds()/60/60)
working_hour =  diff.total_seconds()/60/60
if working_hour >=12:
  print("장시간 근무")
else:
  print("정상 근무")
```
