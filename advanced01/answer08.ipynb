### 1) 오늘 날짜를 `"YYYY-MM-DD HH:MM:SS"` 형식으로 출력하시오.

초(second)를 제거하여 “분”까지만 출력하는 옵션도 추가하시오.

```python
from datetime import datetime
now = datetime.now()
print(now.strftime("%Y-%m-%d %H:%M:%S"))
print(now.strftime("%Y-%m-%d %H:%M"))
```

### 2) `"2025-12-10"` 문자열을 datetime 객체로 변환한 뒤, 요일 이름(예: "Wednesday")을 출력하시오.

```python
from datetime import datetime
str_now = "2025-12-10"
now = datetime.strptime(str_now,"%Y-%m-%d")
weekday = now.strftime("%A")
kor_weekdays = ["월요일","화요일","수요일","목요일","금요일","토요일","일요일",]
print(f"{now}의 요일은 {weekday}")
print(f"{now}의 한글 요일은 {kor_weekdays[now.weekday()]}")
```

### 3) 현재 날짜 기준, 30일 후 날짜가 주말인지 여부를 출력하시오.

```python
from datetime import datetime, timedelta
today = datetime.now().date()
after_30 = today+timedelta(days=30)
#print(type(after_30))
#print(after_30.weekday())
is_weekend = after_30.weekday() >= 5  #월 0 ~ 토5 일6
print(f"{today}에서 30일 뒤는 {after_30}은 주말인가요?{is_weekend}")
```

### 4) `"2025-12-10"` 기준으로 1주일 후, 1개월 후, 1년 후 날짜를 각각 구하시오.

```python
from datetime import datetime, date, timedelta
base_date =  date(2025,12,10)
after_one_week = base_date+timedelta(weeks=1)
temp_month = base_date.month+1
temp_year = base_date.year
if temp_month > 12:
  temp_month = 1
  temp_year+=1
temp_day = base_date.day
while True:
  try:
    after_one_month = date(temp_year,temp_month,temp_day)
    break
  except ValueError:
    temp_day-=1

try:
  after_one_year =  base_date.replace(year=base_date.year+1)
except ValueError:
  after_one_year =  base_date.replace(year=base_date.year+1, day=28)

print(f"기준 날짜 : {base_date}")
print(f"1주일 후 : {after_one_week}")
print(f"1달 후 : {after_one_month}")
print(f"1년 후 : {after_one_year}")
```

```python
#pip install python-dateutil
from datetime import date
from dateutil.relativedelta import relativedelta

base_date =  date(2025,12,10)


# datetime은 days,weeks는 잘 동작함 months, years는 동작안함 
after_one_week = base_date+relativedelta(weeks=+1)
after_one_month = base_date+relativedelta(month=+1)
after_one_year = base_date+relativedelta(years=+1)

print(f"기준 날짜 : {base_date}")
print(f"1주일 후 : {after_one_week}")
print(f"1달 후 : {after_one_month}")
print(f"1년 후 : {after_one_year}")
```
