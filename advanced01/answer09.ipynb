### 1) 오늘 날짜 폴더 자동 생성

**설명:**

오늘 날짜를 기준으로 `logs/YYYY/MM/DD` 형식의 폴더를 자동으로 생성하는 코드를 작성하시오.

예: 2025-12-11 → `logs/2025/12/11` 폴더 생성

```python
from datetime import date
import os
from pathlib import Path

today = date(2025,3,3)
date_path = today.strftime("%Y/%m/%d")
print(today)
#folder = os.path.join("logs",str(today.year),str(today.month),str(today.day))
folder = os.path.join("logs",date_path)  # / 를 문자열로 인식  윈도우에서는 잘 동작
os.makedirs(folder,exist_ok=True)
```

```python
from datetime import date
import os
from pathlib import Path

today = date(2025,3,3)
date_path = today.strftime("%Y/%m/%d")
folder = Path("logs") / date_path  # Path는 / 를 운영시스템에 따라 바꿔준다.
folder.mkdir(parents=True,exist_ok=True)
print(folder)
```

### 2) 특정 확장자만 모아서 리스트 출력

현재 작업 디렉토리에서 **확장자가 `.txt`인 파일만** 찾아서 리스트로 출력하시오.

```python
import os

#print(os.listdir(".")) #os.listdir(".")  현재폴더 기준으로 모든 걸 다 보여줌

# txt_files = [f for f in os.listdir(".") if f.endswith(".txt") and os.path.isfile(f)]
txt_files = []
for f in os.listdir("."):
  if f.endswith(".txt") and os.path.isfile(f):
    txt_files.append(f)
print(txt_files)

#문자열 반환
```

```python
from pathlib import Path
#txt_files = [f for f in Path(".").iterdir() if f.is_file() and f.suffix==".txt"]
txt_files = []
for f in Path(".").iterdir():
  if f.is_file() and f.suffix==".txt":
    txt_files.append(f)
print(txt_files)
#Path 객첼를 반환
```

### 3) 특정 폴더 안의 파일 개수 세기

경로 `"현재"` 폴더 안에 있는 **파일 개수**를 출력하는 코드를 작성하시오.

(하위 폴더는 포함하지 않음)

```python
import os

#print(os.listdir(".")) #os.listdir(".")  현재폴더 기준으로 모든 걸 다 보여줌

# txt_files = [f for f in os.listdir(".") if f.endswith(".txt") and os.path.isfile(f)]
txt_files = []
for f in os.listdir("."):
  if os.path.isfile(f):
    txt_files.append(f)
print(len(txt_files))
```

```python
from pathlib import Path
#txt_files = [f for f in Path(".").iterdir() if f.is_file() and f.suffix==".txt"]
txt_files = []
for f in Path(".").iterdir():
  if f.is_file():
    txt_files.append(f)
print(len(txt_files))
count = sum(1 for f in Path(".").iterdir() if f.is_file())
print(count)
#Path 객첼를 반환
```

### 4) 하위 모든 디렉토리까지 `.py` 파일 찾기

현재 디렉토리 기준, **모든 하위 폴더를 재귀적으로 탐색**하여

`.py` 확장자를 가진 파일의 **절대 경로**를 전부 출력하시오.

```python
import os
for root,dirs,files in os.walk("."):  # 재귀 탐색 즉 모든 폴더를 검색
  # print(root)
  # print(dirs)
  # print(files)
  for file in files:
    if file.endswith(".py") :
      print(os.path.abspath(os.path.join(root,file)))
```
