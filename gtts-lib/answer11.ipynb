## 1) 전체 문장 번역 → 한 개 음성 파일 만들기

### 문제

- `예제.txt` 파일을 읽는다.
- 전체 내용을 **영어 → 한국어**로 번역한다.
- 번역된 내용을 하나의 문자열로 합친다.
- 음성 파일 `result_01.mp3`를 생성한다.

````python
from pathlib import Path
from googletrans import Translator
from gtts import gTTS
import asyncio
# current_dir = Path(__file__).resolve().parent
# 쥬피터 노트북에는 __file__이 없다/.
current_dir = Path.cwd()
exam_dir =  current_dir / "example"
exam01 = exam_dir / "exam01.txt"
ouput_dir = current_dir / "exam01.mp3"
async def trans():
  translator = Translator()
  with open(exam01,"r", encoding="utf-8") as f:
    msg = f.read()
  result = await translator.translate(msg,dest="ko",src="en")
  print(result)
  gtts = gTTS(text=result.text,lang="ko")
  gtts.save(str(ouput_dir))


await trans()  
#  쥬피터 노트북은 이미 event llop가 만들져 있다.
#  따라서 asyncio.run(trans()) 를 쓸 필요가 없다.
````

## 2) 한 줄씩 번역 → 줄별 음성 파일 생성

### 문제

- `예제.txt`를 한 줄씩 읽는다.
- 각 줄을 영어 → 한국어로 번역한다.
- 번역된 문장을 **줄마다 mp3 파일**로 저장한다.
    - `line_01.mp3`
    - `line_02.mp3`
- 빈 줄은 무시한다.

### 예제.txt

```
Good morning.
How are you today?

I like programming.

```

````python
from pathlib import Path
from googletrans import Translator
from gtts import gTTS
import asyncio

current_dir = Path.cwd()
exam_dir =  current_dir / "example"
exam01 = exam_dir / "exam02.txt"
ouput_dir = current_dir / "output"
ouput_dir.mkdir(parents=True,exist_ok=True)

async def trans():
  translator = Translator()
  with open(exam01,"r", encoding="utf-8") as f:
    lines = f.readlines()
  idx = 1
  for line in lines:
    line = line.strip()
    if line:
      result = await translator.translate(line,dest="ko",src="en")
      print(result.text)
      output_mp3 = f"{ouput_dir}/{idx:02d}.mp3"
      gtts = gTTS(text=result.text,lang="ko")
      gtts.save(output_mp3)
      idx+=1
await trans()
````
