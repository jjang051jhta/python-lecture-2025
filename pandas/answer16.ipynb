### 20) duration의 결측치를 전체 데이터의 평균 러닝타임으로 채우세요.

```python
import pandas as pd
df = pd.read_csv("./korean_netflix.csv")
avg_duration = df.duration.mean()
print(avg_duration)
df.duration = df.duration.fillna(avg_duration)
print(df)
```
