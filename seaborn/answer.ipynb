### 1) `tips` 데이터셋을 사용하여

요일(`day`)별 팁(`tip`) 금액의 분포를 **히스토그램 + KDE**로 시각화하시오.

(요일별 색상 구분)

**Hint**

- `sns.histplot()`
- `hue="day"`
- `stat="density"`, `kde=True`

```python
import seaborn as sns  #seaborn의 별칭은 sns 사람이름  
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
import seaborn as sns  #seaborn의 별칭은 sns 사람이름  
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

tips =  sns.load_dataset("tips")
#print(tips.head())
plt.figure(figsize=(8,5))
sns.histplot(data=tips, x="tip",hue="day", bins=30, stat="count",
             kde=False, palette="Set2", alpha=0.6, multiple="stack")
#multiple layer, dodge, fill, stack
# stack은 쌓아서 나옴 주로 count와 조합
# layer는 겹쳐서 나옴 주로 density와 조합
plt.title("요일별 팁 분포 히스토그램")
```

### 2) `tips` 데이터셋에서

식사 시간(`time`)에 따라 **총 계산 금액(`total_bill`)과 팁(`tip`)의 관계**를

산점도로 나타내고, **회귀선 포함/미포함 그래프를 비교**하시오.

**Hint**

- `sns.regplot()`
- `fit_reg=True / False`
- subplot 사용

```python
import seaborn as sns  #seaborn의 별칭은 sns 사람이름  
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

tips =  sns.load_dataset("tips")
fig = plt.figure(figsize=(14,5))
ax01 = fig.add_subplot(1,2,1)
sns.regplot(data=tips, y="tip",x="total_bill",ax=ax01, 
                scatter_kws={"color":"red", "alpha":0.5}, 
                line_kws={"color":"blue", "lw":2},
                fit_reg=True
                )
ax01.set_title("산점도: total_bill과 tip의 관계 회귀선 포함")
ax02 = fig.add_subplot(1,2,2)
sns.regplot(data=tips, y="tip", x="total_bill",ax=ax02, 
                scatter_kws={"color":"red", "alpha":0.5}, 
                line_kws={"color":"blue", "lw":2} ,
                fit_reg=False
                )
fig.suptitle("Seaborn의 regplot으로 산점도와 회귀선 그리기", y=1.05)
```

### 3) `tips` 데이터셋을 이용해

성별(`sex`)에 따른 **팁 비율(`tip / total_bill`)의 분포**를 비교하시오.

**Hint**

- 파생변수 생성
- `sns.boxplot()` 또는 `sns.histplot()`

```python
import seaborn as sns  #seaborn의 별칭은 sns 사람이름  
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

tips =  sns.load_dataset("tips")

tips["tip_ratio"] =  tips["tip"] / tips["total_bill"]*100
#print(tips)
plt.figure(figsize=(8,5))
sns.histplot(data=tips, x="tip_ratio",hue="sex", bins=30, stat="density",
             multiple="dodge",palette="Set1", alpha=0.6)
```

```python
import seaborn as sns  #seaborn의 별칭은 sns 사람이름  
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

tips =  sns.load_dataset("tips")

tips["tip_ratio"] =  tips["tip"] / tips["total_bill"]*100
#print(tips)
fig = plt.figure(figsize=(8,5))
ax01 = fig.add_subplot(1,2,1)
ax02 = fig.add_subplot(1,2,2)

sns.barplot(data=tips, x="sex",y="tip_ratio", palette="Set2", ax=ax01)
sns.boxplot(data=tips, x="sex",y="tip_ratio", palette="Set2", ax=ax02)
ax01.set_title("성별 팁 비율 Barplot")
ax02.set_title("성별 팁 비율 Boxplot")
ax01.set_xlabel("성별")
ax01.set_ylabel("팁 비율(%)")
ax02.set_xlabel("성별") 
ax02.set_ylabel("팁 비율(%)")
ax01.set_title("성별 팁 비율 Barplot")
ax02.set_title("성별 팁 비율 Boxplot")
fig.suptitle("성별에 따른 팁 비율 비교", y=1.05)
# plt.title("성별 팁 비율 Boxplot")
# plt.xlabel("성별")
# plt.ylabel("팁 비율(%)")
```

### 4) `titanic` 데이터셋에서

객실 등급(`pclass`)별 **생존자 비율**을 막대그래프로 나타내시오.

(y축은 percent 기준)

**Hint**

- `groupby()` + `mean()`
- `sns.barplot()`
- `stat="percent"` 아님 (직접 계산)

```python
import seaborn as sns  #seaborn의 별칭은 sns 사람이름  
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic =  sns.load_dataset("titanic")
#print(titanic.head())
rate = titanic.groupby("pclass")["survived"].mean().reset_index()
#print(rate)
rate["survived_rate_percent"] = rate["survived"] * 100  #백분율로 변환
plt.figure(figsize=(8,5))
sns.barplot(data=rate, x="pclass", y="survived_rate_percent", palette="Set3")
plt.title("객실 등급별 생존율")
plt.xlabel("객실 등급")
plt.ylabel("생존율(%)")
```

### 5) `titanic` 데이터셋을 사용하여

성별(`sex`)과 생존 여부(`survived`) 간의 관계를 **countplot**으로 표현하시오.

**Hint**

- `sns.countplot()`
- `hue="survived"`

```python
import seaborn as sns  #seaborn의 별칭은 sns 사람이름  
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

titanic =  sns.load_dataset("titanic")
print(titanic.head())

plt.figure(figsize=(8,5))
ax01 = sns.countplot(data=titanic, x="sex", hue="survived", palette="Set1")
ax01.legend(title="생존 여부", labels=["사망","생존"])
plt.xlabel("성별")
plt.ylabel("승객 수")
plt.title("성별 생존자 수 비교")
```
